```python
import pandas as pd
from datetime import datetime as dt
```

```python
df = pd.read_csv('Austin_MetroBike_Trips.csv')
df
```

```python
df2 = df[['Membership Type','Checkout Date','Checkout Kiosk','Return Kiosk','Trip Duration Minutes']]
dfclean = df2.dropna(how='any')
dfclean
dfclean['Membership Type'].value_counts()[:10].sum()
```

```python
s1 = df['Membership Type'].value_counts()[:10]
mt = s1.index.values
mt
```

```python
def convertDatetoMonth(string):
    dto = dt.strptime(string,'%m/%d/%Y').date().month
    return dto
```

```python
mem_mask = df['Membership Type'].isin(mt)
top10 = df[mem_mask][['Membership Type','Checkout Date','Checkout Kiosk', 'Return Kiosk', 'Trip Duration Minutes']]
# top10['monthofdate']=top10['Checkout Date'].dt.month
top10['MonthInt'] = top10['Checkout Date'].map(convertDatetoMonth)
```

```python
top10[:10]
```

```python
pv = top10[['Membership Type', 'Checkout Kiosk', 'MonthInt', 'Trip Duration Minutes']].groupby(['Membership Type', 'Checkout Kiosk', 'MonthInt']).count()
```

```python
len(top10['Checkout Kiosk'].value_counts()).max()
```

```python
dfmerged = top10.merge(latlong,left_on='Checkout Kiosk',right_on='Station')
dfmerged2 = dfmerged.dropna()
dfmerged2
```
